```python
# Running a single speaker model
# Open the file in read mode
with open('/content/drive/MyDrive/Projects/TTS/file.txt', 'r') as file:
  # Read the contents of the file into a list of strings
  file_contents = file.readlines()
print("Read "+str(len(file_contents))+" lines from the file")


# num_lines = file_contents.count()
OUTPUT_PATH = "Spoken_Output_1.wav"

n=0
printme = ""
for line in file_contents:
  n=n+1
  printme+=  re.sub(r'[^a-zA-Z0-9,.!? \n]','',line);
  
  if n%20 == 19:
    OUTPUT_PATH = "Spoken_Output_"+str(n)+".wav"

    # Init TTS with the target model name
    tts = TTS(model_name="tts_models/en/ljspeech/tacotron2-DCA", progress_bar=False, gpu=False)
    # Run TTS
    tts.tts_to_file(text=printme, file_path=OUTPUT_PATH)

    print("Generating Section "+str(n) )
    printme=""
    # printme = "-------------\n"

#process the last part of the file
OUTPUT_PATH = "Spoken_Output_"+str(n)+".wav"

# Init TTS with the target model name
tts = TTS(model_name="tts_models/en/ljspeech/tacotron2-DCA", progress_bar=True, gpu=False)
# Run TTS
tts.tts_to_file(text=printme, file_path=OUTPUT_PATH)
```
